```python
import pandas as pd
```

```python
# Load the extracted table
extracted_table = pd.read_csv('extracted_table.csv', index_col=[0,1])
extracted_table.head()
```

```python
def extract_from_nmrc(selected_value, match_string, year):
    """
    Extract values from HOSP10_YYYY_nmrc.csv
    
    Parameters:
    selected_value: tuple from extracted_table index (col1, col2), where col2 maps to first column in nmrc
    match_string: string that matches the second column in nmrc.csv (e.g. 'A000000')
    year: the year for the nmrc file (e.g. 2020)
    
    Returns:
    DataFrame of matching rows from nmrc.csv
    """
    file_path = f'data/source_data/HOSP10FY{year}/HOSP10_{year}_nmrc.csv'
    df = pd.read_csv(file_path, header=None)
    # selected_value is (col1, col2), use col2 for matching nmrc col0
    matching_rows = df[(df[0] == selected_value[1]) & (df[1] == match_string)]
    return matching_rows
```

```python
# Example usage
# Select a tuple from extracted_table index, e.g. (2, 454148)
# And a match string, e.g. 'A000000'
# And a year, e.g. 2020
result = extract_from_nmrc((2, 454148), 'A000000', 2020)
result
```
